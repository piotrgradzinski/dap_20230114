```python
import pandas as pd
url = 'https://raw.githubusercontent.com/piotrgradzinski/dap_20230114/main/day_6_pgg/matches.csv'
matches = pd.read_csv(url, parse_dates=['date'])
matches
```

```python
matches.home_team.unique()
```

```python
team_a = 'Legia Warszawa'
team_b = 'Wisła Kraków'
matches[
    (
        ( (matches.home_team == team_a) & (matches.away_team == team_b) )
        |
        ( (matches.home_team == team_b) & (matches.away_team == team_a) )
    )
].sort_values('date')
```

```python
team_a = 'Legia Warszawa'
team_b = 'Wisła Kraków'
matches[
    (
        (
            ( (matches.home_team == team_a) & (matches.away_team == team_b) )
            |
            ( (matches.home_team == team_b) & (matches.away_team == team_a) )
        )
        &
        ( (matches.date.dt.year >= 2010) & (matches.date.dt.year <= 2015) )
    )
].sort_values('date')
```

```python
team_a = 'Legia Warszawa'
team_b = 'Wisła Kraków'
matches[
    (
        (
            ( (matches.home_team == team_a) & (matches.away_team == team_b) )
            |
            ( (matches.home_team == team_b) & (matches.away_team == team_a) )
        )
        &
        ( matches.date.dt.year.isin(range(2010, 2016)) )
    )
].sort_values('date')
```

```python
team_a = 'Legia Warszawa'
team_b = 'Wisła Kraków'
matches[
    (
        (
            ( (matches.home_team == team_a) & (matches.away_team == team_b) )
            |
            ( (matches.home_team == team_b) & (matches.away_team == team_a) )
        )
        &
        ( matches.date.dt.year.isin([2010, 2011, 2012, 2013, 2014, 2015]) )
    )
].sort_values('date')
```
