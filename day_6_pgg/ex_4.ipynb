I want to see 10 matches with the biggest difference in scored goals.

- create a new column which will store the difference between scored goals between the teams
- `abs()` to get the absolute value of the substraction as we may have negative values
- sort our DataFrame by newly created column in a descending mode

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/piotrgradzinski/dap_20230114/main/day_6_pgg/matches.csv'
matches = pd.read_csv(url, parse_dates=['date'])
matches
```

```python
matches['goals_diff'] = abs(matches['home_team_goal'] - matches['away_team_goal'])
matches
```

```python
matches.sort_values('goals_diff', ascending=False).head(10)  # 10 first rows
```

We have a problem if we have more results with 8 goals difference...

```python
matches.sort_values('goals_diff', ascending=False).head(20)
```

```python
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.nlargest.html
matches.sort_values('goals_diff', ascending=False).nlargest(n=10, columns=['goals_diff'], keep='all')
```
