# Data modifications

We have several dedicated methods to change the data within DataFrame or Series objects:
- [.replace()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html) - will allow to change specified value into a different one
- [.fillna()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html) - used to fill/replace [missing values](https://pandas.pydata.org/docs/user_guide/missing_data.html)
- [.apply()]() - allows us to apply any function (either Python, Pandas or our own function) to the data

We can use regular expression within `.replace()`. We have to enable them by adding `regex=True` named argument.

By default `replace` is finding and replacing whole values, not fragments of them. To make that happen we can enable regular expression and then we will be able to change fragments of texts.

````python
df.replace('Egg', 'Oeuf', regex=True)
````

If we are working with regular expression in Python code, especially providing patterns, then it's best to use r-string or raw strings: `r''`. This means that Python will not try to change any strings inside of thje string (\/).

````python
df.replace(r'Mc(\w+)', r'\1', regex=True)
````

## `apply`

Applies a function to each value from a DataFrame (row) or Series.

```
Calories                 New series
300  -> times_10(300) -> 3000
250  -> times_10(250) -> 2500
370  -> times_10(370) -> 3700
```

````python
def times_10(number):
    return number * 10
````

````python
times_10(5)
````

When we want to say that `apply` should use a particular function (either our function defined by `def` keyword or any built-in function) we just provide the name of the function `apply(times_10)` and I'm NOT adding `()` after the name of the function `apply(times_10())` (incorrect), becuase adding `()` means that I'm executing, I'm running the function.


````python
df.Calories.apply(times_10)
````

We can do the same with `lambda` functions. 

````python
df.Calories.apply(lambda number: number * 10)
````

## How we can extract information about grams from a `Serving_Size` column?

````python
import re

def extract_grams(serving_size: str) -> float:
    matches = re.search(r'\((\d+) g\)', serving_size)
    if matches:
        return float(matches[1])
    else:
        return 0.0

df.Serving_Size, df.Serving_Size.apply(extract_grams)
````

````python
df['Serving_Size_Grams'] = df.Serving_Size.apply(extract_grams)
df
````
