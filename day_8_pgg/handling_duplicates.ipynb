```python
emps_selection = emps[['city', 'department_name', 'job_title']]
emps_selection
```

For handling duplicates we can use [`.drop_duplicates()`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop_duplicates.html). If I have several rows with the same values in each column I can use `drop_duplicates` to get rid of duplicated rows and leave only one row.

```python
emps_selection.drop_duplicates()
```

By default `.drop_duplicates` treats whole row as one value and removes repeating values/rows. If we want to change the behaviour and narrow the colums that are use for determining if we have duplicates or not we can use `subset` argument.

```python
emps.drop_duplicates(subset='city')  # I have one employee from one city
```

```python
emps.drop_duplicates(subset=['department_name', 'job_title'])
```

```python
# Lets take one employee from each city with the biggest salary
emps.sort_values('salary', ascending=False).drop_duplicates(subset='city')
```

When dropping duplicates we can decide which values should be kept by using `keep` option and we can use:
- first
- last
- `False` -> all duplicated values will be removed

```python
emps.drop_duplicates(subset='city', keep='first')
```

```python
emps.drop_duplicates(subset='city', keep='last')
```

```python
emps[emps.city == 'London']
```

```python
emps[emps.city == 'Seattle']
```

```python
# Remove all the duplicates and leave nothing behind
emps.drop_duplicates(subset='city', keep=False)
```
